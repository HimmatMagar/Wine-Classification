```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
x = df.drop(columns='quality')
y = df['quality']

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
lr_model = ElasticNet()
```

```python
param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1, 10],
    'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]
}
```

```python
grid = GridSearchCV(lr_model, param_grid, cv=5, scoring='r2')
grid.fit(x_train, y_train)
```

```python
print("Best alpha:", grid.best_params_['alpha'])
print("Best l1_ratio:", grid.best_params_['l1_ratio'])
print("Best R2 score:", grid.best_score_)
```

```python
rfc = RandomForestRegressor()
```

```python
param_grid = {
    "n_estimators": [50, 100, 150],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"]
}
```

```python
grids = GridSearchCV(rfc, param_grid, cv=5, scoring='r2')
grids.fit(x_train, y_train)
```

```python
print("Best parameter:", grids.best_params_)
print("Best R2 score:", grids.best_score_)
```

```python
model = ElasticNet(alpha=0.001, l1_ratio=0.1)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_val)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print("mae", mae)
print("r2", r2)
```
